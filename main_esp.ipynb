## 8. Escribe una funci贸n que devuelva la moda de una lista, es decir: el elemento que aparece m谩s veces.

`NOTE: No se debe usar count... `

```python
def mode_counter(arr):
    num_dict = {}
    for num in arr:
        if num in num_dict:
            num_dict[num] += 1
        else:
            num_dict[num] = 1
    i = 0
    max_num = 0
    for n in num_dict:
        if num_dict.get(n) > i:
            i = num_dict.get(n)
            max_num = n
    return max_num
```
